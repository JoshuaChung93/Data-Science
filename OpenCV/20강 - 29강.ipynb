```python
# 20강 - 캡쳐 및 녹화

import datetime
import cv2

capture = cv2.VideoCapture("so-pissed.gif")
fourcc = cv2.VideoWriter_fourcc(*'XVID') # 디지털 미디어 포멧 코드 생성, 인코딩 방식 설정
record = False # 녹화 유무

while True:
    if(capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT)): # 현재 프레임과 동영상의 총 프레임 수를 비교해서 맞다면 파일을 염
        capture.open("so-pissed.gif") 
        
    ret, frame = capture.read() # 프레임 읽기 메서드를 이용하여 카메라의 상태 및 프레임을 받아옵니다
    cv2.imshow("so-pissed.gif", frame)
    
    now = datetime.datetime.now().strftime("%d_%H_%M_%S") # 파일의 제목 설정, 날짜_시간_분_초의 형식으로 제목 설정
    key = cv2.waitKey(33) # 33ms마다 현재 눌러진 키보드 키값 갱신
    
    if key == 27: # 키값 판단, 27 = esc, 26 = Ctrl + Z, 24 = Ctrl + X, 3 = Ctrl + C
        break
    elif key == 26:
        print('캡쳐')
        cv2.imwrite("C:/" + str(now) + ".png", frame)
    elif key == 24:
        print("녹화 시작")
        record = True
        video = cv2.VideoWriter("C:/" + str(now) + ".avi", fourcc, 20.0, (frame.shape[1], frame.shape[0]))
    elif key == 3:
        print("녹화 중지")
        record = False
        video.release()
        
    if record == True:
        print("녹화 중..")
        video.write(frame)
        
capture.release()
cv2.destroyAllWindows()
```

```python
# 21강 - 윤곽선 검출

src = cv2.imread("contour.webp", cv2.IMREAD_COLOR)

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY) # 하얀색 객체 검출, 배경은 검은색, 검출하려는 물체는 하얀색으로 변형
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY) # 이진화 처리 후, 반전시켜 검출하려는 물체를 하얀색의 성질을 띄도록 변환
binary = cv2.bitwise_not(binary)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE) # 이진화 이미지에서 contour 검색, 반환값으로 contour, hierarchy 반환

for i in range(len(contours)): # 반복문을 사용하여 검출된 윤곽선을 그려 해당 윤곽선의 계층 구조 표시
    cv2.drawContours(src, [contours[i]], 0, (0, 0, 255), 2)
    cv2.putText(src, str(i), tuple(contours[i][0][0]), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0,255,0), 1)
    print(i, hierarchy[0][i])
    cv2.imshow("src", src)
    cv2.waitKey(0)


cv2.destroyAllWindows()
```

```python
# 22강 - 다각형 근사
import cv2

src = cv2.imread("phone.webp", cv2.IMREAD_COLOR)

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
binary = cv2.bitwise_not(binary)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_TC89_KCOS)

for contour in contours:
    epsilon = cv2.arcLength(contour, True) * 0.02 # 반복문 사용해 색인값과 하위 윤곽선 정보로 반복, 근사치 정확도를 계산하기 위해 윤곽선 전체 길이의 2%로 활용
    approx_poly = cv2.approxPolyDP(contour, epsilon, True) # 윤곽선들의 윤곽점들로 근사해 근사 다각형으로 반환, epsilon 이 낮을수록, 근사를 더 적게해 원본 윤곽과 비슷해짐
    
    for approx in approx_poly:
        cv2.circle(src, tuple(approx[0]), 3, (255,0,0), -1) # 반복문을 통해 근사 다각형을 반복해 근사점을 이미지 위에 표시
        
cv2.imshow("src", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 23강 - 코너 검출
import cv2
import numpy as np

src = cv2.imread("cup.webp")
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 5, blockSize=3, useHarrisDetector=True, k=0.03) # 코너 검출. 코너 품질 0.01에 최댓값 100, 최소거리 값 5

for i in range(corners.shape[0]):
    pt = (int(corners[i, 0, 0]), int(corners[i, 0, 1]))
    cv2.circle(dst, pt, 3, (0, 0, 255), 2) # tuple([0]) 센터 값 문제 있음

cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 24강 - 블록 껍질
src = cv2.imread("convex.webp")
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

for i in contours:
    hull = cv2.convexHull(i, clockwise=True) # 윤곽선에서 블록 껍질 검출
    cv2.drawContours(dst, [hull], 0, (0,0,255), 2)
    
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 25강 - 모멘트
src = cv2.imread("convex.webp")
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

for i in contours: 
    M = cv2.moments(i, False) # 윤곽선에서 모멘트 계산
    cX = int(M['m10']/M['m00'])
    cY = int(M['m01'])
    
    cv2.circle(dst, (cX, cY), 3, (255,0,0), -1)
    cv2.drawContours(dst, [i], 0, (0,0,255), 2)
    
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 26강 - 모폴로지 변환

import numpy as np
import cv2

src = cv2.imread('zebra.webp')

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (9,9)) # 구조요소 생성 
dilate = cv2.dilate(src, kernel, anchor=(-1,-1), iterations=5) # 생성된 구조 요소를 활용해 모폴로지 변환 적용. 팽창 연산 진행; 밝은 영역 커지게
erode = cv2.erode(src, kernel, anchor=(-1,-1), iterations=5) # 침식 연산 진행; 어두운 영역 커지게

dst = np.concatenate((src, dilate,erode), axis=1) # 원본 이미지, 팽창 결과, 침식 결과를 하나의 이미지로 연결

height, width, channel = dst.shape # 사이즈 축소
dstsize = (width/10, height/10)
dst2 = cv2.pyrDown(dst, dstsize, borderType=cv2.BORDER_DEFAULT)

cv2.imshow('dst2', dst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
